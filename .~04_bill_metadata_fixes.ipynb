## Add Past Initiators from KNS_BillHistoryInitiator

```python
init = pd.read_csv('Bill_Initiator_2020_10_13.csv.gz')
init.head()
```

```python
ihist = pd.read_csv('Knesset_initiator_history_2020_10_13.csv.gz')
ihist.head()
```

```python
ihist[(~(ihist.PersonID.isna())) & (ihist.FirstName.isna()) ]
```

```python
init = (init.append(ihist.loc[(~(ihist.PersonID.isna())), ['BillID', 'PrivateNumber', 'IsInitiator', 'PersonID',
                                                      'FirstName', 'LastName', 'GenderDesc']])
        .drop_duplicates(subset=['BillID', 'PersonID']))
```

```python
init['Ordinal'] = init.Ordinal.fillna(999).astype(int)
```

```python
init = init.sort_values(['BillID', 'Ordinal'])
```

```python
names_init = init.groupby('BillID', sort=False).apply(lambda x: ','.join((x.FirstName + ' ' + x.LastName).tolist()))
names_init.head()
```

```python
genders_init = init.groupby('BillID').apply(lambda x: ','.join((x.GenderDesc).tolist()))
genders_init.head()
```

```python
ordinals_init = init.groupby('BillID').apply(lambda x: ','.join((x.Ordinal.astype(str)).tolist()))
ordinals_init.head()
```

```python
removed_init = init[init.Ordinal==999].groupby('BillID').size()
removed_init.head()
```

```python
numbers_init = init.groupby(['BillID', 'GenderDesc']).size().unstack().fillna(0).astype(int).rename(columns={'זכר': 'initiators_male', 'נקבה': 'initiators_female'})
numbers_init['initiators_total'] = numbers_init.sum(axis=1)
numbers_init.head()
```

```python
df_with_mb = df.append(mb)[df.columns]
```

```python
df_with_mb.drop(['initiators', 'female_init', 'male_init', 'num_init'], axis=1, inplace=True)
df_with_mb.columns
```

```python
df_with_mb['initiator_names'] = df_with_mb.BillID.map(names_init)
df_with_mb['initiator_genders'] = df_with_mb.BillID.map(genders_init)
df_with_mb['initiator_ordinals'] = df_with_mb.BillID.map(ordinals_init)
df_with_mb['initiator_total_female'] = df_with_mb.BillID.map(numbers_init.initiators_female)
df_with_mb['initiator_total_male'] = df_with_mb.BillID.map(numbers_init.initiators_male)
df_with_mb['initiator_total'] = df_with_mb.BillID.map(numbers_init.initiators_total)
df_with_mb['initiator_removed'] = (df_with_mb.BillID.map(removed_init)).fillna(0).astype(int)
df_with_mb.tail().T
```
